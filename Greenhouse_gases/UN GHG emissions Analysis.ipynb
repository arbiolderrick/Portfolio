### Compare and contrast ghg emissions amongst the UN's top 10 highest GDP countries

```python
#### Top 10 includes: U.S., Germany, UK, France, Italy, Canada, Australia, Spain, Turkey, Netherlands see full 2017 list 
#### https://data.worldbank.org/indicator/NY.GDP.MKTP.CD?view=map&year_high_desc=true
```

```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv file
ghg_df = pd.read_csv("greenhouse_gas_inventory_data_data.csv")
ghg_df.tail()
```

```python
#list unique category values for filtering later
categories = ghg_df.category.unique()
categories
```

```python
ghg_df.country_or_area.unique()
```

### United States 

```python
#U.S., Germany, UK, France, Italy, Canada, Australia, Spain, Turkey, Netherlands

#sort by US
US1_df = ghg_df.loc[ghg_df['country_or_area']=='United States of America']

#sort year values ascending values; reset index/drop old one
US1_df.sort_values(by='year',ascending=True).reset_index(drop=True).tail(25)
```

```python
#how many entries per type
US1_df['category'].value_counts()
```

```python
# 1990-2014 US Average emission by category 

US_ghg_emiss = US1_df.groupby('category')['value'].mean()
US_ghg_emiss
```

```python
US_ghg_emiss[5:]
```

### Germany

```python
#sort by Germany
Ger2_df = ghg_df.loc[ghg_df['country_or_area']=='Germany']

#sort year values ascending values; reset index/drop old one
Ger2_df.sort_values(by='year',ascending=True).reset_index(drop=True).tail(25)
```

```python
# number of German entries per category 
Ger2_df['category'].value_counts()
```

```python
#Average Germany emissions by category

Ger_ghg_emiss = Ger2_df.groupby('category')['value'].mean()
Ger_ghg_emiss
```

# Breakdown of average emissions from Top 5 GDP UN countries

```python
#labels for x-axis; need to relabel later
x_axis = ["CO2_emissions_without_LULUCF", 
          "GHGs_emissions_including_indirect_co2_without_LULUCF", 
          "GHGs_emissions_without_LULUCF",
          "HFCs_emissions",
          "CH4_emissions_without_LULUCF",
          "NF3_emissions",
          "N2O_emissions_without_LULUCF",
          "PFCs_emissions",
          "SF6_emissions",
          "Unspecified_mix_of_HFCs_and_PFCs_emissions"]        

#x_axis = categories

#Top 5 GDP UN countries' mean GHG emissions
y_axis1 = US_ghg_emiss
y_axis2 = Ger_ghg_emiss
y_axis3 = UK_ghg_emiss
y_axis4 = Fr_ghg_emiss
y_axis5 = Ity_ghg_emiss

#United States GHG emissions
plt.bar(x_axis, y_axis1, width = 0.7, label="United States")
plt.xticks(rotation=90)


plt.title("US GHG emissions")
plt.xlabel("Category")
plt.ylabel("Average GHG Emitted (kt co2)")
#
plt.legend(title=["Legend"])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.grid(True)
plt.savefig('UnitedStates_GHG.png')
plt.show()
```
